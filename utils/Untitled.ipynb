```python
import pandas as pd
import matplotlib.pyplot as plt
import json
from dateutil import parser
```

```python
data = pd.DataFrame()
with open('/Users/mypolopony/Downloads/log.log','r') as log:
    for line in log.readlines():
        comma = line.index(',')
        date = parser.parse(line[:comma])
        row = line[comma+1:].replace("'",'"').replace('\n','').strip()
        rowdf = pd.DataFrame(json.loads(row), index=[date])
        data = data.append(rowdf)
```

```python
data.plot()
plt.show()
```
